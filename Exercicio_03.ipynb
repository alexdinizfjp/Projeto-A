```python
# Questão 01

matriculas = []

matriculas.append(101)
matriculas.append(102)
matriculas.append(103)
matriculas.append(104)

print("Matrículas armazenadas:")
indice = 0
while indice < len(matriculas):
    print(f"Posição {indice}: Matrícula {matriculas[indice]}")
    indice += 1

print(f"\nTotal de elementos inseridos: {len(matriculas)}")
```

```python
# Questão 02

estoque = []

estoque.append(45)
estoque.append(33)
estoque.append(27)
estoque.append(89)
estoque.append(56)
estoque.append(72)

print("--- Relatório de Estoque ---")
indice = 0
while indice < len(estoque):
    print(f"Posição [ {indice} ] -> Código do Produto: {estoque[indice]}")
    indice += 1

print(f"\nTotal de itens registrados: {len(estoque)}")

# O sistema funcionou com o novo código adicionado
```

```python
# Questão 03


clientes = [11, 22, 33, 44, 55]
procurado = 33
encontrada = -1
indice = 0

while indice < len(clientes):
    if clientes[indice] == procurado:
        encontrada = indice
        break
    indice += 1

if encontrada != -1:
    print(f"O cliente com código {procurado} foi encontrado!")
    print(f"Posição na lista (índice): {encontrada}")
else:
    print(f"Cliente {procurado} não está presente na lista.")

```

```python
# Questão 04


clientes = [10, 20, 30, 40, 50]

print(f"Lista antes da exclusão: {clientes}")

remover = 30

if remover in clientes:
    clientes.remove(remover)
    print(f"\nSucesso: O cliente {remover} foi excluído.")
else:
    print(f"\nErro: Código {remover} não encontrado.")

print(f"Lista atualizada: {clientes}")
```

```python
# Questão 05


pacientes = [5, 6, 7, 8, 9]

print(f"Fila de espera inicial: {pacientes}")

senha_desistencia = 7

if senha_desistencia in pacientes:
    pacientes.remove(senha_desistencia)
    print(f"\nNotificação: O paciente com a senha {senha_desistencia} desistiu e foi removido.")
else:
    print(f"\nErro: Senha {senha_desistencia} não encontrada na fila.")

print(f"Fila de espera atualizada: {pacientes}")
```

```python
# Questão 06


CAPACIDADE_MAXIMA = 6
reservas = [101, 103, 107, 110, 115]

print(f"Estado atual das reservas: {reservas}")
print(f"Vagas disponíveis: {CAPACIDADE_MAXIMA - len(reservas)}")

reserva_procurada = 108

encontrada = False

if reserva_procurada in reservas:
    encontrada = True
    posicao = reservas.index(reserva_procurada)
else:
    encontrada = False

print(f"\n--- Resultado da Pesquisa ---")
if encontrada:
    print(f"A reserva {reserva_procurada} JÁ está cadastrada na posição {posicao}.")
else:
    print(f"A reserva {reserva_procurada} NÃO foi encontrada no sistema.")
    print("O número está disponível para um novo cadastro.")
```

```python
# Questão 07


pedidos = [2001, 2002, 2003, 2004, 2005]

print(f"Lista de pedidos original: {pedidos}")

pedido_cancelado = 2002

if pedido_cancelado in pedidos:
    pedidos.remove(pedido_cancelado)
    print(f"\nConfirmação: O pedido {pedido_cancelado} foi cancelado e removido do fluxo.")
else:
    print(f"\nErro: O pedido {pedido_cancelado} não foi encontrado no sistema.")

print(f"Lista de pedidos atualizada: {pedidos}")
```

```python
# Questão 08


estacionamento = []

estacionamento.append(1234)
estacionamento.append(5678)
estacionamento.append(9101)
estacionamento.append(1213)

print("--- Controle de Vagas Ocupadas ---")
for i in range(len(estacionamento)):
    print(f"Vaga {i + 1}: Placa {estacionamento[i]}")

capacidade_maxima = 4
total_ocupado = len(estacionamento)

print(f"\nStatus: {total_ocupado}/{capacidade_maxima} vagas ocupadas.")

if total_ocupado == capacidade_maxima:
    print("ALERTA: O estacionamento está LOTADO.")
```

```python
# Questão 09


notas = [70, 85, 90, 75, 60]

nota_procurada = 75

posicao_encontrada = -1
indice = 0

while indice < len(notas):
    if notas[indice] == nota_procurada:
        posicao_encontrada = indice
        break
    indice += 1

print(f"--- Sistema de Consulta de Notas ---")
if posicao_encontrada != -1:
    print(f"A nota {nota_procurada} foi localizada com sucesso!")
    print(f"Posição na lista (índice): {posicao_encontrada}")
else:
    print(f"A nota {nota_procurada} não consta no sistema.")
```
