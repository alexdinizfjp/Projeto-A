```python
# Questão 01

import math

class Circulo:
    def __init__(self, raio):
        self.raio = raio

    def calcular_area(self):
        area = math.pi * (self.raio**2)
        return area

raio_usuario = float(input("Digite o valor do raio aqui: "))
meu_circulo = Circulo(raio_usuario)
resultado = meu_circulo.calcular_area()

print(f"Um circulo com raio {meu_circulo.raio} tem uma área de: {resultado:.2f}")
```

```python
# Questão 02

class Livro:
    def __init__(self, titulo, autor):
        self.titulo = titulo
        self.autor = autor

    def detalhes(self):
        return f"Livro: {self.titulo}, autor: {self.autor}"

livro1 = Livro("Violetas na Janela", "Vera Lúcia Marinzeck")
livro2 = Livro("O Evangelho Segundo o Espiritismo", "Allan Kardec")

print(livro1.detalhes())
print(livro2.detalhes())
```

```python
# Questão 03

class Retangulo:
  def __init__(self, base, altura):
    self.base = base
    self.altura = altura

  def calcular_area(self):
    area = self.base * self.altura
    return area

meu_retangulo = Retangulo(5, 10)
resultado = meu_retangulo.calcular_area()

print(f"A base é {meu_retangulo.base}")
print(f"A altura é {meu_retangulo.altura}")
print(f"A área total do retangulo é: {resultado}")
```

```python
# Questão 04

class ContaBancaria:
    def __init__(self, saldo, titular):
        self.saldo = saldo
        self.titular = titular

    def depositar (self, valor):
        if valor > 0:
            self.saldo += valor
            print(f"Depósito de R$ {valor:.2f} realizado com sucesso!")

        else:
            print("Erro: O valor do depósito deve ser positivo.")

    def sacar (self, valor):
        if valor <= 0:
            print("Erro: Saldo insuficiente!.")
        else:
            self.saldo -= valor
            print(f"Saque de R$ {valor:.2f} realizado. Novo saldo: R$ {self.saldo:.2f}")

minha_conta = ContaBancaria(1000, "Alex Diniz")

minha_conta.depositar(500)
minha_conta.sacar(200)
minha_conta.sacar(2000)
```

```python
# Questão 05

class Pessoa:
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def falar(self):
        print(f"Olá! Meu nome é {self.nome} e eu tenho {self.idade} anos.")

pessoa1 = Pessoa("Alex Diniz", 48)
pessoa2 = Pessoa("Diana", 50)

pessoa1.falar()
pessoa2.falar()
```

```python
# Questão 06

class Produto:
    def __init__(self, nome, preco, quantidade):
        self.nome = nome
        self.preco = preco
        self.quantidade = quantidade

    def calcular_total(self):
        total = self.preco * self.quantidade
        return total

p1 = Produto("Caderno Universitário", 25.50, 10)

valor_estoque = p1.calcular_total()

print(f"Produto: {p1.nome}")
print(f"Preço Unitário: R$ {p1.preco:.2f}")
print(f"Quantidade: {p1.quantidade}")
print(f"Valor Total: R$ {valor_estoque:.2f}")
```

```python
# Questão 07

class Carro:

    def __init__(self, marca, modelo, ano):
        self.marca = marca
        self.modelo = modelo
        self.ano = ano

    def detalhes(self):
        return f"Carro: {self.marca} {self.modelo}, Ano: {self.ano}"

carro1 = Carro("Toyota", "Corolla", 2024)
carro2 = Carro("BYD", "Dolphin", 2025)

print(carro1.detalhes())
print(carro2.detalhes())
```

```python
# Questão 08

class Aluno:

    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas

    def calcular_media(self):
        media_final = sum(self.notas) / 3
        return media_final

notas_do_alex = [8.5, 9.0, 10.0]
aluno1 = Aluno("Alex Diniz", notas_do_alex)

media_final = aluno1.calcular_media()

print(f"Aluno: {aluno1.nome}")
print(f"Notas: {aluno1.notas}")
print(f"Média Final: {media_final:.2f}")
```

```python
# Questão 09

class Triangulo:

    def __init__(self, lado1, lado2, lado3):
        self.lado1 = lado1
        self.lado2 = lado2
        self.lado3 = lado3

    def calcular_perimetro(self):
        perimetro = self.lado1 + self.lado2 + self.lado3
        return perimetro

meu_triangulo = Triangulo(5, 7, 10)

resultado = meu_triangulo.calcular_perimetro()

print(f"Lados do triângulo: {meu_triangulo.lado1}, {meu_triangulo.lado2}, {meu_triangulo.lado3}")
print(f"O perímetro do triângulo é: {resultado}")
```

```python
# Questão 10

class Funcionario:
    def __init__(self, nome, salario, cargo):
        self.nome = nome
        self.salario = salario
        self.cargo = cargo

    def aumentar_salario(self, percentual):
        if percentual > 0:
            aumento = self.salario * (percentual / 100)
            self.salario += aumento
            print(f"Aumento de {percentual}% aplicado com sucesso!")
        else:
            print("Erro: O percentual de aumento deve ser positivo.")

    def mostrar_dados(self):
        return f"Nome: {self.nome}, Salário: R$ {self.salario:.2f}, Cargo: {self.cargo}"

f1 = Funcionario("Alex Diniz", 5000.00, "Coordenador Técnico")

print("Dados Iniciais:")
print(f1.mostrar_dados())

f1.aumentar_salario(10)

print("\nDados Atualizados:")
print(f1.mostrar_dados())
```
