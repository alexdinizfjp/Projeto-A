```python
# Questão 04

numeros = []
continuar = 's'

while continuar.lower() == 's':
  num = int(input("Digite um número: "))
  numeros.append(num)
  continuar = input("Deseja adicionar outro número? (s/n): ")

  if len(numeros) > 0:
    maior = numeros[0]
    menor = numeros[0]

    indice = 1
    while indice < len(numeros):
      if numeros[indice] > maior:
        maior = numeros[indice]

      if numeros[indice] < menor:
        menor = numeros[indice]

      indice += 1

    print(f"\nLista final: {numeros}")
    print(f"O maior número é: {maior}")
    print(f"O menor número é: {menor}")
```

```python
# Questão 05

numeros = []
continuar = "s"

while continuar.lower() == 's':
    num = int(input("Digite um número: "))
    numeros.append(num)
    continuar = input("Deseja adicionar outro número? (s/n): ")

soma_pares = 0
contador_pares = 0

for num in numeros:
    if num % 2 == 0:
        soma_pares += num
        contador_pares += 1

if contador_pares > 0:
    media = soma_pares / contador_pares
    print(f"\nForam encontrados {contador_pares} números pares")
    print(f"A média dos números pares é: {media:.2f}")
```
