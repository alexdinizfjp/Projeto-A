```python
# Questão 01

def selection_sort(vetor):
    n = len(vetor)

    for i in range(n):
        indice_minimo = i

        for j in range(i + 1, n):
            if vetor[j] < vetor[indice_minimo]:
                indice_minimo = j

        vetor[i], vetor[indice_minimo] = vetor[indice_minimo], vetor[i]

    return vetor

meu_vetor = [64, 25, 12, 22, 11]
print(f"Vetor original: {meu_vetor}")

vetor_ordenado = selection_sort(meu_vetor)
print(f"Vetor ordenado: {vetor_ordenado}")
```

```python
# Questão 02

def ordenar_vetor(vetor, ordem="crescente"):
    n = len(vetor)

    for i in range(n):
        for j in range(0, n - i - 1):

            if ordem == "crescente":
                condicao = vetor[j] > vetor[j + 1]
            else:
                condicao = vetor[j] < vetor[j + 1]

            if condicao:
                vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]

    return vetor

numeros = [45, 12, 89, 33, 27]

print("Original:", numeros)
print("Crescente:", ordenar_vetor(numeros.copy(), "crescente"))
print("Decrescente:", ordenar_vetor(numeros.copy(), "decrescente"))
```

```python
# Questão 03

def encontrar_extremos(vetor):
    if not vetor:
        return None, None

    maior = vetor[0]
    menor = vetor[0]

    for i in range(1, len(vetor)):

        if vetor[i] > maior:
            maior = vetor[i]

        if vetor[i] < menor:
            menor = vetor[i]

    return maior, menor

dados_estoque = [45, 12, 89, 33, 27, 56, 8]
valor_max, valor_min = encontrar_extremos(dados_estoque)

print(f"Vetor de análise: {dados_estoque}")
print(f"Maior valor encontrado: {valor_max}")
print(f"Menor valor encontrado: {valor_min}")
```

```python
# Questão 04

def encontrar_segundo_menor(vetor):

    if len(set(vetor)) < 2:
        return None

    menor = float('inf')
    segundo_menor = float('inf')

    for num in vetor:
        if num < menor:
            segundo_menor = menor
            menor = num

        elif num < segundo_menor and num != menor:
            segundo_menor = num

    return segundo_menor

notas_teste = [10, 20, 10, 15, 30, 15]
resultado = encontrar_segundo_menor(notas_teste)

print(f"Vetor de análise: {notas_teste}")
print(f"O segundo menor número distinto é: {resultado}")
```

```python
# Questão 05

def remover_duplicados(vetor):

    vetor_unico = []

    for elemento in vetor:
        ja_existe = False

        for unico in vetor_unico:
            if elemento == unico:
                ja_existe = True
                break

        if not ja_existe:
            vetor_unico.append(elemento)

    return vetor_unico

codigos_produtos = [101, 102, 101, 103, 104, 102, 105]
resultado = remover_duplicados(codigos_produtos)

print(f"Lista original com duplicados: {codigos_produtos}")
print(f"Lista saneada (apenas únicos): {resultado}")


```

```python
# Questão 06

def processar_vetor(vetor):
    n = len(vetor)

    for i in range(n):
        for j in range(0, n - i - 1):

            if vetor[j] < vetor[j + 1]:
                vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]

    pares = 0
    impares = 0

    for num in vetor:
        if num % 2 == 0:
            pares += 1
        else:
            impares += 1

    return vetor, pares, impares

dados = [15, 42, 7, 18, 90, 33, 10]
vetor_ordenado, total_pares, total_impares = processar_vetor(dados)

print(f"Vetor em Ordem Decrescente: {vetor_ordenado}")
print(f"Quantidade de Números Pares: {total_pares}")
print(f"Quantidade de Números Ímpares: {total_impares}")
```

```python
# Questão 07

def encontrar_terceiro_maior(vetor):

    unicos = []
    for num in vetor:
        if num not in unicos:
            unicos.append(num)

    if len(unicos) < 3:
        return "Erro: O vetor não possui pelo menos 3 valores distintos."

    n = len(unicos)
    for i in range(n):
        indice_max = i
        for j in range(i + 1, n):
            if unicos[j] > unicos[indice_max]:
                indice_max = j
        unicos[i], unicos[indice_max] = unicos[indice_max], unicos[i]

    return unicos[2]

notas_ti = [95, 80, 95, 100, 85, 100, 70]

resultado = encontrar_terceiro_maior(notas_ti)
print(f"Vetor de análise: {notas_ti}")
print(f"O terceiro maior valor distinto é: {resultado}")
```

```python
# Questão 08

def calcular_mediana_impar(vetor):

    if not vetor:
        return None

    vetor_ordenado = sorted(vetor)

    n = len(vetor_ordenado)
    indice_meio = n // 2

    return vetor_ordenado[indice_meio]

dados_estoque = [45, 12, 89, 33, 27]

mediana = calcular_mediana_impar(dados_estoque)

print(f"Vetor Original: {dados_estoque}")
print(f"Vetor Ordenado: {sorted(dados_estoque)}")
print(f"A Mediana é: {mediana}")
```
